# Section 8. Introduction to Matplotlib

## 42. Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11) #0 to 5, grab 11 points
y = x**2
```

```python
# Two method to create plot 
# Method 1: functional method 

plt.plot(x,y,)
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.title('TITLE')
```

```python
# Put multi-plots on the same canvas 
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Method 2: Object oriented API method (better way)

fig = plt.figure() # Create a figure object first

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # Then call the method and 
# add a set of aces and passed in a list 

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Large Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
# Add subplot using OOP method 

fig,axes = plt.subplots(nrows=1,ncols=2) #one row, 2 columns = 2 plot

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
# Figure size, DPI(dots per inch)

fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2)) # 8 inches by 2 inches 

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# Save a figure 

fig.savefig('my_picture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc= )
```

```python
# Plot Appearance 

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange', # we can set colors as string ,
        lw=2, # set width of line 
       alpha = 0.5,   # set transparency of line 
       ls = '--',      # set line style
       marker = 'o',   # mark all pints
       markersize = 10,  # specify marker size 
       markerfacecolor = 'red',
       markeredgewidth=3,
       markeredgecolor = 'green')                              
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange', 
        lw=2, 
       alpha = 0.5,   
       ls = '--') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

# Section 9. Seaborn

## 48. Distribution plots 

```python
import seaborn as sns
```

```python
%matplotlib inline  # we can see our visualization insdie of the notebook
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Distribution plot: show the distribution of a univariate set of 
# observations 

sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
# Jointplot: 2 distribution plots and one scatter plot 
# kind will change the type of plot inside (defualt is scatterplot)

sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
# Pairplot: pair plots for all hte numerical column values 
# Efficient to quickly visualize our data 

sns.pairplot(tips, 
             hue='sex',  # hue could let us add categorical column 
            palette = 'coolwarm') # choose certain colors 
```

```python
# Rugplot: draw a dash mark for every points on the uniform or unique 
# variant distribution 

sns.rugplot(tips['total_bill'])
```

```python
# KDE plot - kernel density estimation plot 

sns.kdeplot(tips['total_bill'])
```

## 49. Categorical Plots 

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
# Bar plot: general plot that allows us aggregate the categorical data 
# base off of some function and by default that is the mean 

sns.barplot(x='sex', # x is categorical data 
            y='total_bill', # y is numerical data 
            data=tips, 
            estimator=np.std) # statistical function to estimate within each 
                        # categorical bin 
```

```python
# Count plot:

sns.countplot(x='sex', data=tips)
```

```python
# Box plot anad violin plot: show the distribution of categorical data 

sns.boxplot(x='day', #categorical data 
            y='total_bill',
           data = tips, # numerical data - try to see the distribution of total bill/day
           hue = 'smoker') # now we can compare not only day by day behavior
                           # for total bill but also by whether or not 
                           # this person is a smoker 
```

```python
# Violin plot, shows more info than boxplot

sns.violinplot(x='day',y='total_bill', data=tips, hue='sex',
              split=True) # split into two,
```

```python
# Strip plot 

sns.stripplot(x='day',y='total_bill',data=tips, hue='sex',
              jitter=True) # add noise, can see the density better 
```

```python
# Swarm plot: similar to strip plot, but the points are adjusted so 
# that they don't overlap 
# give us better representation of the distribution of values 
# It is the combination of strip plot and violin plot 
# DRAWBACK: not scale well to very large numbers 

sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# Factor plot 

sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

## 50. Matrix plots 

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
# we have to convert to matrix form first 
tc=tips.corr()
tc
```

```python
# heatmap 
sns.heatmap(tc, 
            annot=True) # annotate the actual values 
        
```

```python
flights
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='magma', # we can see dark to light 
            linecolor='white', # line between each cell 
            linewidths =1)
```

```python
# Cluster map: has hierarchal clusters and try to cluster columns 
# and rows together based on their similarity 

sns.clustermap(fp)
```

## 51. Grid

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
# PairGrid
g = sns.PairGrid(iris) # we can have different types of plot together 
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# FacetGrid
g = sns.FacetGrid(data=tips,
                 col='time',
                 row='smoker')

g.map(plt.scatter,'total_bill','tip') #total bill vs.tip
```
